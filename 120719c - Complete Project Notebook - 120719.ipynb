### Q1 - How long on average is the EMS response time? How does it compare to national/local guidelines?

As noted previously, in EMS response, seconds matter. Thus, our first question centers on average response time for the SFOEMS. This is the time it takes the SFOEMS to get to a scene after being dispatched (i.e., it is the difference between ArrivalTm and AlarmTm). We are interested in comparing SFO data to national guidelines, which recommend - for 90% of calls - response times of 5 minutes or less (Moore-Merrell 2019) as well as local (SFO) guidelines, which recommend 10 minutes or less (SFOFD 2019b). Thus, in order to conduct our analysis, we need to first "bucket" our data into the requisite categories (i.e., zero seconds*; more than zero seconds through 300 seconds [5 minutes]; more than 300 seconds [5 minutes] through 600 seconds [10 minutes]; more than 600 seconds [10 minutes]). To this end, we felt a bar chart comparison best.

*In reviewing our data, we found that a large number of calls had response times of zero seconds. To us, this situation did not sound feasible for such a large number of calls (i.e., while it's possible that an EMS unit "came upon" an incident in progress or a citizen with an emergency "walked into" a firehouse, both of which would result in a response time of 0, the sheer number of zeros was too high for these explanations to account, entirely, for all of the zeros we found). On further investigation and consultation with an expert in the field (i.e., Lt. William J. Massey of the Wallingford FD), we confirmed our suspicion. According to Lt. Massey, the call reporting system that most FDs use ("EMS Charts") works in such a way that if a dispatcher inadvertently "double-clicks" on the entry for AlarmTm (the time of the alarm), then that entry gets repeated in the box for ArrivalTm (the time the first unit arrives on scene). Lt. Massey suggested that virtually all of the zeros would be from such erroneous data entry. Thus, for purposes of analysis, we segregated the calls showing response times of zero from the rest of the calls, by bucketing them.

References:

Moore-Merrell, L. 2019. Understanding and Measuring Fire Department Response Times. _Lexipol_ (blog). Jul 17. https://www.lexipol.com/resources/blog/understanding-and-measuring-fire-department-response-times/

San Francisco Fire Department (SFOFD). 2019b. _Ambulance Response to Life-Threatening Emergencies_. Accessed: Dec 4. https://sfgov.org/scorecards/public-safety/ambulance-response-life-treatening-emergencies

```python
# Code to create a new column ('Response_Mins') converting 'ResponseTm' to minutes to facilitate comparisons to national/SFO benchmarks
SFOEMS['Response_Mins']=SFOEMS['ResponseTm'].apply(lambda x: x/60).round(2)
SFOEMS.head()
```

```python
##Code to Produce "Buckets" for SFOEMS Response Times (as converted to minutes)

# to create buckets for minutes taken for bar chart
min_bucket = [-.1, 0, 5, 10, 500]
min_taken = pd.cut(SFOEMS['Response_Mins'], min_bucket)
#print(min_taken)

# to get count of values in each bucket
min_taken.value_counts()
```

```python
## Code to Produce Bar Chart for SFOEMS Response Times

ax = min_taken.value_counts().sort_index().plot(kind='bar', grid=False, rot=0, color='cornflowerblue', figsize=(11,9))

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005), fontsize=16, ha='left')

plt.ylim(0,40000)
ax.set_title('Q1 - Frequencies for EMS "Response Times" by Time Bucket', fontsize=20)
ax.set_xlabel('Time Buckets (in Minutes)', fontsize=14)
ax.set_ylabel('Frequency', fontsize=14)
ax.set_xticklabels(('0', '>0 to 5', '>5 to 10', '>10'))
```

### Interpretation of Response Time Data (Q1):

__Overall, the SFOFD is doing a good job in meeting response time standards.__ 
From the above figure, we can see that, for the vast majority of calls, response time is within the response time standards of the SFOFD, which provide for responses within a 10-minute window. That said, however, the SFOFD does not meet the NFPA's (national) response time standard of arriving on scene within five minutes of dispatch for 90% of its calls. Thus, the SFOFD might considering adding staff and/or apparatus in order to improve this metric pursuant to the NFPA standard.

__Nonetheless, as we also can see from the figure above, there remains room for improvement.__ 
First, as noted previously, a large number of calls are potentially mis-entered by dispatchers as having the same arrival time as alarm time; hence, we recommend additional training of the dispatchers to reduce these reporting errors. Similarly, we noted via our descriptive analysis that the maximum response time reported in the dataset is 28853 seconds, which corresponds to just over 8 hours. This finding further suggests the possibility of data input errors on the part of disptachers, which augments the recommendation for additional training of the dispatchers. Finally, the legitimate existence of calls with response times in excess of 10 minutes highlights the need for additional analysis to understand the underlying reasons in order to come up with potential solutions to address the issues.

### Q3 - What is the most frequent emergency type?

Different types of emergencies require different types of apparatus and equipment. For example, fires often require multiple pieces of apparatus (e.g., Engines, Trucks, Rescues, etc.) and equipment (e.g., hoses, ladders, pike poles, etc.) whereas medical emergencies typically require response from fewer pieces/types of apparatus (e.g., response sometimes is limited to just an Ambulance and/or a Rescue) and different types of equipment (e.g., heart montors, defibrillators, etc.). As such, understanding the common emergency types can provide information about the types of apparatus/equipment (and personnel to operate them) that are most needed. To this end, we felt a pie chart best.

```python
# Code to create a new column ('Category_condensed') to improve visualization of emergency subgroups

# Define function to condense "weather", "explosion", "special" emergency subgroups into "Other" for more precise visualization
def new_category(x):
    if x == 'Weather':
        return 'Other'
    elif x == 'Explosion':
        return 'Other'
    elif x == 'Special':
        return 'Other'
    else:
        return x

SFOEMS['Category_condensed']=SFOEMS['Category'].apply(lambda x:new_category(x))

#Print data to check it
SFOEMS.tail()

```

```python
##Code to count how many categories are each type

# count how many categories are each type
cat_count= SFOEMS['Category_condensed'].value_counts()
print(cat_count)
```

```python
## Code to Create Pie Chart to visualize emergency types

# create pie chart visualizing emergency types
#
colors = ['silver', 'blue', 'cornflowerblue', 'crimson', 'green', 'gold', 'lightskyblue']
explode = [0.1, 0, 0, 0, 0, 0, 0]
ax = SFOEMS['Category_condensed'].value_counts().plot(kind='pie', figsize= (10,10), autopct='%1.0f%%',\
                                                    legend=True, labels=None, colors=colors, explode=explode)
ax.set_title('Q3 - Percentage of Emergency Responses By Type', fontsize=20)
ax.set_ylabel('', fontsize=14)
```

## Interpretation of Emergency Type Data (Q3):

__The vast majority of SFOEMS calls are false alarms.__ 
From the above figure, we can see that, almost half of the calls to which the SFOFD reponded  were "False Alarms" (i.e., calls for which there is no good faith belief of a bona fide emergency). These calls are costly for the city as they inflict unnecessary wear and tear on FD personnel and aparatus, can delay response to legitimate calls, and can heighten risks of accidents that can occur when responding to calls. As such, it is in the City's interest to reduce the incidence of false alarms. *** 

As highlighted in Karter's (2013) report, false alarms can be intentional (e.g., when a student "pulls" a fire alarm to cause a distrubance in school) or unintentional (e.g., when a fire alarm sounds due to a malfunction). While lawmakers and law enforcement professionals most likely would be in a position to implement measures that might prevent intentional false alarms, education initiatives can be undertaken to reduce the incidence of unintentional false alarms. For instance, the city of SFO can begin a public education campaign to teach residents about the importance of and ways to keep their smoke and carbon monoxide detectors clean.

__"EMS" (Rescue & Emergency Medical Incidents) are the second-most typical type of call.__ 
Given the relative prevalence of Rescue & Emergency Incidents in SFO, it would behoove the city to plan for more frequent replacement of Ambulances and Rescue Vehicles and also to consider hiring medically-qualified personnel (e.g., Paramedics and/or Emergency Medical Technicians or EMTs) as much as possible. 

*** "Good intent" calls are reportedly separately from "False Alarms" because they arise when emergency services are summoned _in good faith_ when a bona fide emergency was believed to exist, but in fact did not exist.

### Q7 - In what month/season do most EMS calls occur?

As discussed previously, it is important for municipalities to try to allocate resources on the basis of needs. Given this and the fact that SFO is a tourist destination, we thought it might be wise to investigate whether call volumes vary by month/season. Since many people take vacations in the summer, it is possible that call volumes in SFO are higher then, due to the influx of visitors. If this is the case, then summer might not be the best time for fire department personnel to undertake annual equipment maintenance or in-house training - as there would be a greater chance of those activities being interrupted by emergency calls. We utilize Bar Charts to aid in our visualization for purposes of this analysis.


```python
## To conduct our analysis, we first needed to determine "months" in which calls occurred

# extract month from Indcident Date column and add month column
SFOEMS['month'] = pd.DatetimeIndex(SFOEMS['IncidentDate']).month
SFOEMS.head()
```

```python
#To create a dictionary of month number to month name
month_name = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8:
          'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
print(month_name)
```

```python
#To add month_name column to df using append
SFOEMS['Month Name']=SFOEMS['month'].map(month_name)
SFOEMS.tail()
```

```python
#To create a dictionary of month to seasons
season = {1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring', 6: 'Summer', 7: 'Summer', 8:
          'Summer', 9: 'Fall', 10: 'Fall', 11: 'Fall', 12: 'Winter'}
print(season)
```

```python
#To add season column to df using append
SFOEMS['Season']=SFOEMS['month'].map(season)
SFOEMS.tail()
```

```python
# Code to Produce Bar Chart for Call Volume by Month Name (ordered by month)

ax = SFOEMS['Month Name'].value_counts().loc\
[["Jan","Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]]\
.plot(kind='bar', grid=False, rot=45, color='cornflowerblue', figsize=(10,8))

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005), fontsize=10, ha='left')

plt.ylim(4000,6000)
ax.set_title('Q7 - Call Volume by Month', fontsize=20)
ax.set_xlabel('Month', fontsize=14)
ax.set_ylabel('Frequency', fontsize=14)
```

```python
# Code to Produce Bar Chart for Call Volume by Season (ordered by season)

ax = SFOEMS['Season'].value_counts().loc\
[["Winter","Spring", "Summer", "Fall"]]\
.plot(kind='bar', grid=False, rot=45, color='cornflowerblue', figsize=(8,8))

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005), fontsize=10, ha='left')

plt.ylim(15000,16500)
ax.set_title('Q7 - Call Volume by Season', fontsize=20)
ax.set_xlabel('Season', fontsize=14)
ax.set_ylabel('Frequency', fontsize=14)
```

## Interpretation of Data on the Months/Seasons of Highest Call Volumes(Q7):

__Overall, most calls come in during the month of January.__ 
From the above figures, we can see that, January is the month with the highest number of emergency calls. Next are, in descending order, December, July, October and September. By contrast, February shows many fewer calls than the other months. In part, the differences in the number of days in each month may contribute to our observed findings, since all of the top five, with the exception of September have 31 days, whereas February (the month with the lowest call volume) has only 28 days in a regular year (and 29 days in a leap year). Nonetheless, it does appear that there tend to be fewer calls in the period between February and June. Thus, the SFOFD might consider schdeduling annual equipment maintence tasks and in-house training sessions during these months.

__On a seasonal basis, most calls occur in the Fall and Winter.__ 
This finding suggests that although SFO is a tourist destination, call volumes do not seem to be elevated during the summer months. Indeed, except for the Spring, which has a lower call volume, call volumes in the other seasons of the year are very close to 16,000 calls. Consistent with the monthly results discussed above, these results suggest that, if anyting, the SFOFD would want to concentrate annual equipment maintenance and in-house training during the Spring season.

### Q8 - On what day of the week do most EMS calls occur; is there a particular day that has an excess of calls? Do most calls come in on the weekend or weekday?

According to our subject-matter expert, because of the need to provide emergency coverages 24/7, most FDs schedule personnel according to some form of "rotating schedule" that varies work assignments over time. In SFO, FD personnel typically work four 12-hour shifts (i.e., 48 hours) and then have eight 12-hour shifts off (i.e., 96 hours), giving them a rotation over 6 days. Our expert reports working two 10-hour day-time shifts, follwed by two 14-hour over-night shifts, followed by 4 days off, for a total rotation over 8 days. At the same time, most other people follow a 7-day pattern with their schedules. Yet, our expert is not aware of any FD whose schedule rotates over a 7-day period. Thus, in order to provide the SFOFD with insight into the need to consider developing a 7-day schedule rotation that might better match staff with call volume, we undertook an analysis of call volume by day of the week - organizing our results by day of the week to more easily identify whether call volumes peek during weekdays or weekend days. We utilize Bar Charts to aid in our visualization for purposes of this analysis.


```python
# Code to Produce Bar Chart for Call Volume by Day of the Week (ordered by day)
ax = SFOEMS['IncidentDay'].value_counts().loc\
[["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]]\
.plot(kind='bar', grid=False, rot=45, color='cornflowerblue', figsize=(10,8))

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005), fontsize=10, ha='left')

plt.ylim(8000,10000)
ax.set_title('Q8 - Average Call Volume by Weekday/Weekend Day', fontsize=20)
ax.set_xlabel('Time Period', fontsize=14)
ax.set_ylabel('Average', fontsize=14)
```

## Interpretation of Data on the Daily Calls (Q8):

__Overall, most calls occur on Fridays.__ 
From the above figure, we can see that Friday is the day with the highest number of emergency calls, with over 9500 calls logged on Fridays. The other six days of the week each have around 9000 calls logged (i.e., 9000 plus or minus ~150). Thus, in considering its scheduling of FD personnel, SFO might want to consider supplementing personnel on Fridays. Beyond that, from a call volume perspective, there does not appear to be any compelling reason to increase staffing levels at any other time. Accordingly, there is no call-volume basis for SFO to consider instituting a 7-day schedule rotation.

__Overall, weekend call volume is not higher than weekday call volume.__ 
This finding suggests that although SFO is a busy city, call volumes do not seem to be elevated on weekends. Indeed, of the days of the week, Sunday tends to have the lowest call volume. Consistent with the daily results just discussed, these results give no compelling justification to increase staffing levels on the weekend.

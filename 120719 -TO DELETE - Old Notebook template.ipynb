### Q1 - How long on average is the EMS response time? How does it compare to national/local guidelines?

As noted previously, in EMS response, seconds matter. Thus, our first question centers on average response time for the SFOEMS. This is the time it takes the SFOEMS to get to a scene after being dispatched (i.e., it is the difference between ArrivalTm and AlarmTm). We are interested in comparing SFO data to national guidelines, which recommend response times of 5 minutes or less (Moore-Merrell 2019) as well as local (SFO) guidelines, which recommend 10 minutes or less (SFOFD 2019b). Thus, in order to conduct our analysis, we need to first "bucket" our data into the requisite categories (i.e., zero seconds*; more than zero seconds through 300 seconds [5 minutes]; more than 300 seconds [5 minutes] through 600 seconds [10 minutes]; more than 600 seconds [10 minutes]). To this end, we felt a bar chart comparison best.

*In reviewing our data, we found that a large number of calls had response times of zero minutes. To us, this situation did not sound feasible for such a large number of calls (i.e., while it's possible that an EMS unit "came upon" an incident in progress or a citizen with an emergency "walked into" a firehouse, both of which would result in a response time of 0, the sheer number of zeros was too high for these explanations to account, entirely, for all of the zeros we found). On further investigation and consultation with an expert in the field (i.e., Lt. William J. Massey of the Wallingford FD), we confirmed our suspicion. According to Lt. Massey, the call reporting system that most FDs use ("EMS Charts") works in such a way that if a dispatcher inadvertently "double-clicks" on the entry for AlarmTm (the time of the alarm), then that entry gets repeated in the box for ArrivalTm (the time the first unit arrives on scene). Lt. Massey suggested that virtually all of the zeros would be from such erroneous data entry. Thus, for purposes of analysis, we segretated the calls showing response times of zero from the rest of the calls.

References:

Moore-Merrell, L. 2019. Understanding and Measuring Fire Department Response Times. _Lexipol_ (blog). Jul 17. https://www.lexipol.com/resources/blog/understanding-and-measuring-fire-department-response-times/

San Francisco Fire Department (SFOFD). 2019b. _Ambulance Response to Life-Threatening Emergencies_. Accessed: Dec 4. https://sfgov.org/scorecards/public-safety/ambulance-response-life-treatening-emergencies

````python
# Code to create a new column ('Response_Mins') converting 'ResponseTm' to minutes
SFOEMS['Response_Mins']=SFOEMS['ResponseTm'].apply(lambda x: x/60).round(2)
SFOEMS.head()
````

````python
##Code to Produce "Buckets" for SFOEMS Response Times (as converted to minutes)

# to create buckets for minutes taken for bar chart
min_bucket = [-.1, 0, 5, 10, 20]
min_taken = pd.cut(SFOEMS['Response_Mins'], min_bucket)
#print(min_taken)

# to get count of values in each bucket
min_taken.value_counts()
````

````python
## Code to Produce Bar Chart for SFOEMS Response Times

ax = min_taken.value_counts().sort_index().plot(kind='bar', grid=False, rot=45, color='navy', figsize=(11,9))

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005), fontsize=16)

ax.set_title('Q1 - Frequencies of EMS "Response Times"', fontsize=20)
ax.set_xlabel('"Response Time" Buckets (in Minutes)', fontsize=14)
ax.set_ylabel('Frequency', fontsize=14)
````

### Interpretation of Response Time (Q1) Data:

__Overall, we can see from the figure above that the SFOFD is doing a good job in meeting response time standards.__ 
From the above figure, we can see that, for the vast majority of calls, response time is within the response time standards of the NFPA and SFOFD, which provide for responses within 5- and 10-minute windows, respectively. Thus, 

__Nonetheless, as we also can see from the figure above, there remains room for improvement.__ 
First, as noted previously, a large number of calls are mis-entered by dispatchers as having the same arrival time as alarm time; hence, we recommend additional training of the dispatchers to reduce these reporting errors. Similarly, we noted via our descriptive analysis that the maximum response time reported in the dataset is 28853 seconds, which corresponds to just over 8 hours. This finding further suggests the possibility of data input errors on the part of disptachers, which augments the recommendation for additional training of the dispatchers. Finally, the legitimate existence of calls with response times in excess of 10 minutes highlights the need for additional analysis to understand the underlying reasons in order to come up with potential solutions to address the issues.

### Q3 - What is the most frequent emergency type?

Different types of emergencies require different types of apparatus and equipment. For example, fires often require multiple pieces of apparatus (e.g., Engines, Trucks, Rescues, etc.) and equipment (e.g., hoses, ladders, pike poles, etc.) whereas medical emergencies typically require response from fewer pieces/types of apparatus (e.g., response sometimes is limited to just an Ambulance and/or a Rescue) and different types of equipment (e.g., heart montors, defibrillators, etc.). As such, understanding the common emergency types can provide information about the types of apparatus/equipment (and personnel to operate them) that are most needed. To this end, we felt a pie chart best.

````python
# Code to create a new column ('Category_condensed') to improve visualization of emergency subgroups

# Define function to condense "weather", "explosion", "special" emergency subgroups into "Other" for more precise visualization
def new_category(x):
    if x == 'Weather':
        return 'Other'
    elif x == 'Explosion':
        return 'Other'
    elif x == 'Special':
        return 'Other'
    else:
        return x

SFOEMS['Category_condensed']=SFOEMS['Category'].apply(lambda x:new_category(x))

#Print data to check it
SFOEMS.tail()

````

````python
##Code to count how many categories are each type

# count how many categories are each type
cat_count= SFOEMS['Category_condensed'].value_counts()
print(cat_count)
````

````python
## Code to Create Pie Chart to visualize emergency types

# create pie chart visualizing emergency types
ax = SFOEMS['Category_condensed'].value_counts().plot(kind='pie', figsize= (10,10), autopct='%1.0f%%', legend=True, labels=None)
ax.set_title('Q3 - Percentage of Emergencies Responded To By Type', fontsize=20)
ax.set_ylabel('', fontsize=14)
````

### Interpretation of Emergency Type (Q3) Data:

__Overall, we can see from the figure above that the vast majority of calls are false alarms.__ 
From the above figure, we can see that, almost half of the calls to which the SFOFD reponded  were "False Alarms" (i.e., calls for which there is no good faith belief of a bona fide emergency). These calls are costly for the city as they inflict unnecessary wear and tear on FD personnel and aparatus (and can heighten risks of accidents that can occur when responding to calls). As such, it is in the City's interest to reduce the incidence of false alarms. *** 

As highlighted in Karter's (2013) report, false alarms can be intentional (e.g., when a student "pulls" a fire alarm to cause a distrubance in school) or unintentional (e.g., when a fire alarm sounds due to a malfunction). While lawmakers and law enforcement professionals most likely would be in a position to implement measures that might prevent intentional false alarms, education initiatives can be undertaken to reduce the incidence of unintentional false alarms. For instance, the city of SFO can begin a public education campaign to teach residents about the importance of and ways to keep their smoke and carbon monoxide detectors clean.

__As well, as we also can see from the figure above, the second-most typical type of call is "EMS", which are Rescue & Emergency Medical Incidents.__ 
Given the relative prevalence of Rescue & Emergency Incidents in SFO, it would behoove the city to consider hiring medically-qualified personnel (e.g., Paramedics and/or Emergency Medical Technicians or EMTs) as much as possible. 

*** "Good intent" calls are reportedly separately from "False Alarms" because they arise when emergency services are summoned _in good faith_ when a bona fide emergency was believed to exist, but in fact did not exist.

````python
# Code to create a new column ('Response_Mins') converting 'ResponseTm' to minutes
SFOEMS['Response_Mins']=SFOEMS['ResponseTm'].apply(lambda x: x/60).round(2)
SFOEMS.head()
````

````python
# Define function to define and add a variable for response categories
def response_cat(x):
    if x == 0:
        return 'A'
    elif 0 < x <= 5:
        return 'B'
    elif 5 < x <= 10:
        return 'C'
    elif x == 'Special':
        return 'D'
    else:
        return x

SFOEMS['Response_Cat']=SFOEMS['Response_Mins'].apply(lambda x:response_cat(x))

#Print data to check it
SFOEMS.head()
````
